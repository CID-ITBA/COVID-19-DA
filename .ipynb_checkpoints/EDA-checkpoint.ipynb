```python
from sklearn.linear_model import LinearRegression

def get_confirmed(country="Argentina"):
    return np.array(df_confirmed[(df_confirmed["Country/Region"] == country)])[0,4:].astype(int)
def get_recovered(country="Argentina"):
    return np.array(df_recovered[(df_recovered["Country/Region"] == country)])[0,4:].astype(int)
def get_deaths(country="Argentina"):
    return np.array(df_deaths[(df_deaths["Country/Region"] == country)])[0,4:].astype(int)
def get_infected(country="Argentina"):
    return (get_confirmed(country)-get_recovered(country)-get_deaths(country))
def get_countries():
    return df["Country/Region"].unique()
def days_since_first_infected(country="Argentina"):
    return sum(get_confirmed(country)>0)
def get_exp(country="Argentina", from_last=-20, to_last=0, window_size=5):
    exponents = list()
    for idx in range(from_last, to_last-window_size):
        reg=LinearRegression()
        X = get_infected(country)[idx:idx+window_size]
        reg.fit(np.array(range(0,window_size)).reshape(-1,1),np.log(X).reshape(-1,1))
        exponents.append(np.exp(reg.coef_[0]))
    return exponents
def get_dates():
    return df_confirmed.columns[4:]
def plot_stats(country="Argentina"):
    dsfi = days_since_first_infected(country)
    plt.figure(figsize=[18,12])
    plt.plot(get_confirmed(country)[-dsfi:],label="Confirmed")
    plt.plot(get_recovered(country)[-dsfi:],label="Recovered")
    plt.plot(get_deaths(country)[-dsfi:],label="Deceased")
    plt.plot(get_infected(country)[-dsfi:],label="Infected")
    plt.legend()
    plt.show()
```
